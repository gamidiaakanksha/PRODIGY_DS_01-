```python
import pandas as pd
import matplotlib.pyplot as plt
url_path = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/refs/heads/main/Task%201/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(url_path, skiprows=4)
df_2022 = df[['Country Name', '2022']]
df_2022 = df_2022.dropna()
top10countries = df_2022.sort_values(by='2022', ascending=False).head(10)
plt.figure(figsize=(8, 5))
plt.bar(top10countries['Country Name'], top10countries['2022'], color='skyblue')
plt.title('Top 10 Countries by Population (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(7, 4))
plt.hist(df_2022['2022'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Population Across Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url_path = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/refs/heads/main/Task%201/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df_meta = pd.read_csv(url_path)


region_counts = df_meta['Region'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(region_counts.index, region_counts.values, color='violet')
plt.title('Number of Countries per Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
income_counts = df_meta['IncomeGroup'].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(income_counts.index, income_counts.values, color='lightgreen')
plt.title('Number of Countries per Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
